```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Replace 'your_image_url' with the actual URL of the image you want to load
image_url = 'https://api.time.com/wp-content/uploads/2023/09/time100-next-ronald-acuna-jr-2023.jpg'
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))
```

```python
# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Resize the image to 224x224
resized_image = image.resize((224, 224))

# Display the resized image
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
# Convert the image to grayscale
grayscale_image = resized_image.convert("L")
```

```python
# Display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = np.array(resized_image)

# Create 10 random filters
filters = [np.random.randn(3, 3) for _ in range(10)]

# Create a figure to display filters and feature maps
plt.figure(figsize=(15, 5))

# Display each filter and corresponding feature map
for i, filter in enumerate(filters):

    feature_map = cv2.filter2D(image, -1, filter)

    plt.subplot(2, 10, i + 1)
    plt.imshow(filter, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 10, i + 11)
    plt.imshow(feature_map, cmap='gray')
    plt.axis('off')

plt.show()
# This code will create 10 random filters, apply each filter to your resized image, and display both the filters and feature maps in Google Colab. You can adjust the number of filters, their sizes, and other parameters as needed.
```
